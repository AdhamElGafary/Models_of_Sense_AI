```python
import nltk
import spacy
from transformers import pipeline
from nltk.tokenize import sent_tokenize
from collections import Counter
import string

# Download necessary NLTK resources
nltk.download('punkt')

nltk.download('punkt_tab')
# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Hugging Face Summarization Model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def nltk_summarize(text, num_sentences=3):
    """ Extractive summarization using NLTK """
    sentences = sent_tokenize(text)
    word_frequencies = Counter(word.lower().strip(string.punctuation) for word in text.split())
    ranked_sentences = sorted(sentences, key=lambda s: sum(word_frequencies[word.lower()] for word in s.split()), reverse=True)
    return " ".join(ranked_sentences[:num_sentences])

def spacy_summarize(text, num_keywords=5):
    """ Keyword extraction and concise summarization using spaCy """
    doc = nlp(text)
    keywords = [token.text for token in doc if token.is_alpha and not token.is_stop]
    most_common = [word for word, _ in Counter(keywords).most_common(num_keywords)]
    summary = " ".join(most_common)
    return f"Keywords: {summary}"

def huggingface_summarize(text, min_length=50, max_length=150):
    """ Abstractive summarization using Hugging Face Transformers """
    summary = summarizer(text, min_length=min_length, max_length=max_length, do_sample=False)
    return summary[0]['summary_text']

# Example Usage
if __name__ == "__main__":

    sample_text = input("Enter Text to Summarize: ")

    # print("\n[NLTK Summary]:")
    # print(nltk_summarize(sample_text))

    # print("\n[spaCy Keywords]:")
    # print(spacy_summarize(sample_text))

    print("\n[Hugging Face Summary]:")
    print(huggingface_summarize(sample_text))
```
